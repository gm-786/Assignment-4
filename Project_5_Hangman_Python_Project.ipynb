```python
import random

words = ["python", "hangman", "developer", "computer", "keyboard", "laptop", "function"]

stages = [
    """
       +---+
       |   |
           |
           |
           |
           |
    =========
    """,
    """
       +---+
       |   |
       O   |
           |
           |
           |
    =========
    """,
    """
       +---+
       |   |
       O   |
       |   |
           |
           |
    =========
    """,
    """
       +---+
       |   |
       O   |
      /|   |
           |
           |
    =========
    """,
    """
       +---+
       |   |
       O   |
      /|\\  |
           |
           |
    =========
    """,
    """
       +---+
       |   |
       O   |
      /|\\  |
      /    |
           |
    =========
    """,
    """
       +---+
       |   |
       O   |
      /|\\  |
      / \\  |
           |
    =========
    """
]

word = random.choice(words)
guessed_word = ["_"] * len(word)
guessed_letters = []
wrong_attempts = 0
max_attempts = len(stages) - 1

print("Welcome to Hangman!")
print("Guess the word, one letter at a time.")
print(f"You have {max_attempts} wrong guesses allowed.")

while wrong_attempts < max_attempts and "_" in guessed_word:
    print(stages[wrong_attempts])
    print("\nCurrent word:", " ".join(guessed_word))
    guess = input("Guess a letter: ").lower()

    if not guess.isalpha() or len(guess) != 1:
        print(" Please enter a single valid letter.")
        continue

    if guess in guessed_letters:
        print("You already guessed that letter.")
        continue

    guessed_letters.append(guess)

    if guess in word:
        print("Good guess!")
        for i in range(len(word)):
            if word[i] == guess:
                guessed_word[i] = guess
    else:
        print("Wrong guess!")
        wrong_attempts += 1

    print("Guessed letters:", ", ".join(guessed_letters))
    print("Remaining attempts:", max_attempts - wrong_attempts)
    print()

if "_" not in guessed_word:
    print(stages[wrong_attempts])
    print("Congratulations! You guessed the word:", word)
else:
    print(stages[wrong_attempts])
    print("Game Over! The word was:", word)
```
